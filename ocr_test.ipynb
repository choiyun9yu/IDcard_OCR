```python
from paddleocr import PaddleOCR, draw_ocr
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

### 1. 저화질 원본 테스트 (50KB 이하)

```python
ocr = PaddleOCR(use_angle_cls=True, lang='korean') # need to run only once to download and load model into memory
img_path = 'data/img(low).jpg'
result = ocr.ocr(img_path, cls=True)
result = result[0]
for line in result:
    print(line)

image = Image.open(img_path).convert('RGB')
boxes = [line[0] for line in result]
txts = [line[1][0] for line in result]
scores = [line[1][1] for line in result]
im_show = draw_ocr(image, boxes, txts, scores, font_path='data/HMFMPyun.ttf')  # 폰트는 원하는 것으로 설정 가능
im_show = Image.fromarray(im_show)
im_show.save('data/result/img(low).jpg')
im_show
```

## kakao vision API

- 비교적 정확학함
- Input image Max bite : 8MB
- Input image Max size : 4,096 * 4,096 px

- 서비스 종료  
https://devtalk.kakao.com/t/api-notice-end-of-support-for-vision-translation-and-speech-apis/122817

- naver / google은 유료 

```python
# 22년 7월 1일 부로 카카오 서비스 종료 
import json
import time
import cv2
import requests
import sys
from google.colab.patches import cv2_imshow

LIMIT_PX = 1024
LIMIT_BYTE = 1024 * 1024  # 1MB
LIMIT_BOX = 40

# EDA function
def kakao_ocr_resize(image_path: str):
    image = cv2.imread(image_path)
    height, width, _ = image.shape
    
    if LIMIT_PX < height or LIMIT_PX < width:
        ratio = float(LIMIT_PX) / max(height, width)
        image = cv2.resize(image, None, fx=ratio, fy=ratio)
        height, width, _ = height, width, _ = image.shape
        image_path = "{}_resized.jpg".format(image_path)
        cv2.imwrite(image_path, image)
        return image_path
    
    return None

# kakao API requests
def kakao_ocr(image_path: str, appkey: str):
    API_URL = 'https://dapi.kakao.com/v2/vision/text/ocr'
    headers = {'Authorization':'KakaoAK {}'.format(appkey)}
    image = cv2.imread(image_path)
    jpeg_image = cv2.imencode(".jpg", image)[1]
    data=jpeg_image.tobytes()

    return requests.post(API_URL, headers=headers, files={"iamge": data})

# main function
def main():
    if len(sys.argv) != 3:
        print("Please run with args : $ python example.py /path/to/image appkey")
    image_path, appkey = '/content/drive/MyDrive/Colab Notebooks/face_id/data/story.png', 'ce961a6807a472c5eb71902feed43ce7'
    white = [255, 255, 255]
    img=cv2.imread(image_path)
    constant = cv2.copyMakeBorder(img,200,100,100,100, cv2.BORDER_CONSTANT, value=white)
    image_path = '/content/drive/MyDrive/Colab Notebooks/face_id/data/story.png'
    cv2.imwrite(image_path, constant)
    cv2_imshow(constant) 
    time.sleep(2)

    resize_impath = kakao_ocr_resize(image_path)
    if resize_impath is not None:
        image_path = resize_impath
        print("원본 대신 리사이즈된 이미지를 사용합니다.")
        
    output = kakao_ocr(image_path, appkey).json()
    print("[OCR] output:\n{}\n".format(json.dumps(output, sort_keys = True, ensure_ascii=False, indent=2)))


if __name__ == "__main__":
    main()
```
