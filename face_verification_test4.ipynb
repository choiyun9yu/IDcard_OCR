# Deepface를 활용하여 진행

```python
from deepface import DeepFace
import os
# 사용할 모델 종류
models = ['VGG-Face', "Facenet", "Facenet512", "OpenFace",
          "DeepFace", "DeepID", "ArcFace", "Dlib"]

# 평가 방법
metrics = ["cosine", "euclidean", "euclidean_12"]
# distance_metric를 무엇으로 설정하는지에 따라서 확연한 성능차이를 보인다.
```

### Facenet 기본 모델 성능 테스트

#### Actual True Data Test

```python
# 경로 설정
dataset = 'C:/Users/SY/Desktop/AI code/IDcard_OCR/data/dataset'
for i in range(1,101):
    data_path = f'{dataset}/{i}'
    data_path = f'{dataset}/{i}'
    os.chdir(f'{dataset}/{i}')
    file_list=os.listdir()
    face_path = f'{data_path}/{file_list[1]}'
    card_path = f'{data_path}/{file_list[2]}'

    # 두 얼굴쌍 비교
    try: 
        result = DeepFace.verify(img1_path=face_path, img2_path=card_path, # Theshold : 10
                                model_name=models[1],  # Backebone : FaceNet
                                distance_metric=metrics[1])  # Distance : Euclidean
        res = result['verified']
        dist = result['distance']
        print(f'#CASE{i} 결과 : {res}, 거리 : {dist}')
    except:
        print(f'#CASE{i} 얼굴 검출 실패!')
```

#### Actual False Data Test

```python
from random import seed
from random import randint
seed(1)

dataset = 'C:/Users/SY/Desktop/AI code/IDcard_OCR/data/dataset'
for k in range(1,101):
    while True:
        i = randint(1,100)
        j = randint(1,100)
        if i != j : break

    face_path = f'{dataset}/{i}'
    os.chdir(f'{face_path}')
    file_list=os.listdir()
    face_path = f'{face_path}/{file_list[1]}'
    
    card_path = f'{dataset}/{j}'
    os.chdir(f'{card_path}')
    file_list=os.listdir()
    card_path = f'{card_path}/{file_list[2]}'

    try: 
        result = DeepFace.verify(img1_path=face_path, img2_path=card_path, # Theshold : 10
                                model_name=models[1],  # Backebone : FaceNet
                                distance_metric=metrics[1])  # Distance : Euclidean
        res = result['verified']
        dist = result['distance']
        print(f'얼굴사진 {i}번, 신분증 {j}번')
        print(f'#CASE{k} 결과 : {res}, 거리 : {dist}')
    except:
        print(f'#CASE{k} 얼굴 검출 실패!')
```
