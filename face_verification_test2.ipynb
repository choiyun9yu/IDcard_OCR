```python
import cv2
import face_recognition as fr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 비교 이미지 까지 같이 있는 ver
class Face_recognition2():
    #def __init__(self):
    #    import cv2
    #    import face_recognition as fr
    #    import numpy as np
    #    import matplotlib.pyplot as plt


    def verification(self, img1_path, img2_path):
        # 데이터 로드
        img1 = cv2.imread(img1_path)
        img2 = cv2.imread(img2_path)

        # 이미지 색상변환
        img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
        img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

        # 얼굴 영역 검출
        img_face1 = fr.face_locations(img1)
        img_face2 = fr.face_locations(img2)

        # Crop
        img1_crop = img1[img_face1[0][0]:img_face1[0][2], img_face1[0][3]:img_face1[0][1]] 
        img2_crop = img2[img_face2[0][0]:img_face2[0][2], img_face2[0][3]:img_face2[0][1]] 

        # 얼굴 특징 분석
        encode_img1 = fr.face_encodings(img1_crop)[0]
        encode_img2 = fr.face_encodings(img2_crop)[0]

        # 얼굴 인식
        res = fr.compare_faces([encode_img1], encode_img2)

        # 거리 분석
        dist = fr.face_distance([encode_img1], encode_img2)

        # 얼굴 영역 그리기
        if res[0] == 1:
            cv2.rectangle(img1, (img_face1[0][3], img_face1[0][0]),
                         (img_face1[0][1], img_face1[0][2]),
                         (0, 255, 0), 30)  # (rgb코드), 두께

            cv2.rectangle(img2, (img_face2[0][3], img_face2[0][0]),
                                (img_face2[0][1], img_face2[0][2]),
                                (0, 255, 0), 30)  
        else: 
            cv2.rectangle(img1, (img_face1[0][3], img_face1[0][0]),
                        (img_face1[0][1], img_face1[0][2]),
                        (255, 0, 0), 30)  # (rgb코드), 두께

            cv2.rectangle(img2, (img_face2[0][3], img_face2[0][0]),
                                (img_face2[0][1], img_face2[0][2]),
                                (255, 0, 0), 30)  

        # 비교 영역 출력 
        plt.subplot(1, 2, 1)
        plt.imshow(img2)
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(img1)
        plt.axis('off')

        plt.show()
        
        print(f'동일인 여부 : {res}', f'거리 : {dist}')

        return res, dist
```
